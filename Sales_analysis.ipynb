```python
df[['Order Date', "Ship Date"]] = df[['Order Date', "Ship Date"]].apply(pd.to_datetime, format='%d/%m/%Y')
```

```python
df.sort_values(by=['Order Date', "Ship Date"], inplace=True)
```

Calculate the average delivery time (Ship Date - Order Date) for orders.

```python
Time_taken_deliver= df["Ship Date"] - df['Order Date']
df["Time_taken_deliver"] = Time_taken_deliver
```

```python
df
```

```python
df["Time_taken_deliver"].mean()
```

Analyze sales trends over time and identify seasonality patterns.

```python
df.set_index("Order Date", inplace=True)
```

```python
Date_Sales = df.resample("D")["Sales"].sum()
Month_Sales = df.resample("M")["Sales"].sum()
```

```python
Date_Sales.idxmax() #In this Date Maximum sales has gained
Month_Sales.idxmax() #In This Month Maximum sales has gained
```

```python
#These are the rows of the maximum sales accured
df.loc[Date_Sales.idxmax()]
df.loc[Month_Sales.idxmax()]
```

```python
plt.figure(figsize=(16, 6))
plt.title("Sales over Time")
plt.plot(Month_Sales.index, Month_Sales.values)
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```
